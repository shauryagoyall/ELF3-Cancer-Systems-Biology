```python
import pandas as pd
from scipy import stats

gene_networks =["elf3"]
replicates = ["r1","r2","r3"]
stability_states = [1,2,3,4,5]
num_of_nodes = 8
genes = ["SLUG", "MIR", "ZEB",  "ERA66", "ERA36","ELF3","PDL1","CDH1"]
topo ="elf3"

for network in gene_networks:
    for replicate in replicates:
        for stability_state in stability_states:
            
            file_name = "./"+network+"/"+replicate+"/"+topo+"_solution_"+str(stability_state)+".dat"
            
            if stability_state == 1:
                
                df1 = pd.read_csv(file_name, sep="\t", header = None)
                df1 = df1.drop(columns = [0,1])
                
                #print(df1)
                
            if stability_state == 2:
                
                df2 = pd.read_csv(file_name, sep="\t", header = None)
                df2 = df2.drop(columns = [0,1])
                
                #print(df2)
                
            if stability_state == 3:
                
                df3 = pd.read_csv(file_name, sep="\t", header = None)
                df3 = df3.drop(columns = [0,1])
                
                #print(df3)
            if stability_state == 4:
                
                df7 = pd.read_csv(file_name, sep="\t", header = None)
                df7 = df7.drop(columns = [0,1])
            
            if stability_state == 5:
                
                df21 = pd.read_csv(file_name, sep="\t", header = None)
                df21 = df21.drop(columns = [0,1])
                
        #changing column index of latter half of bi stable data
        df4 = df2[ range( num_of_nodes+2, ((num_of_nodes*2)+2) ) ].copy(deep = True)
        df4.set_axis(  range(2,num_of_nodes+2) , axis=1, inplace=True)
        
        #changing column index of middle 1/3 of tri stable data
        df5 = df3[ range( num_of_nodes+2, ((num_of_nodes*2)+2) ) ].copy(deep = True)
        df5.set_axis( range(2,num_of_nodes+2) , axis=1, inplace=True)
        
        #changing column index of latter 1/3 of tri stable data
        df6 = df3[ range( ((num_of_nodes*2)+2), ((num_of_nodes*3)+2) ) ].copy(deep = True)
        df6.set_axis( range(2,num_of_nodes+2) , axis=1, inplace=True)
        
        #changing column index of 2/4th of quad stable data
        df8 = df7[ range( num_of_nodes+2, ((num_of_nodes*2)+2) ) ].copy(deep = True)
        df8.set_axis( range(2,num_of_nodes+2) , axis=1, inplace=True)
        
        #changing column index of 3/4thof quad stable data
        df9 = df7[ range( ((num_of_nodes*2)+2), ((num_of_nodes*3)+2) ) ].copy(deep = True)
        df9.set_axis( range(2,num_of_nodes+2) , axis=1, inplace=True)
        
        #changing column index of 4/4thof quad stable data
        df10 = df7[ range( ((num_of_nodes*3)+2), ((num_of_nodes*4)+2) ) ].copy(deep = True)
        df10.set_axis( range(2,num_of_nodes+2) , axis=1, inplace=True)
        #####################################################################
        
        #####################################################################
        #changing column index of 2/5th of penta stable data
        df22 = df21[ range( num_of_nodes+2, ((num_of_nodes*2)+2) ) ].copy(deep = True)
        df22.set_axis( range(2,num_of_nodes+2) , axis=1, inplace=True)
        
        #changing column index of 3/5thof penta stable data
        df23 = df21[ range( ((num_of_nodes*2)+2), ((num_of_nodes*3)+2) ) ].copy(deep = True)
        df23.set_axis( range(2,num_of_nodes+2) , axis=1, inplace=True)
        
        #changing column index of 4/5thof penta stable data
        df24 = df21[ range( ((num_of_nodes*3)+2), ((num_of_nodes*4)+2) ) ].copy(deep = True)
        df24.set_axis( range(2,num_of_nodes+2) , axis=1, inplace=True)
        
        #changing column index of 5/5thof penta stable data
        df25 = df21[ range( ((num_of_nodes*4)+2), ((num_of_nodes*5)+2) ) ].copy(deep = True)
        df25.set_axis( range(2,num_of_nodes+2) , axis=1, inplace=True)

        #concatenate all the dataframes one below the other
        df_new = pd.concat([ df1, \
                            df2[range(2,num_of_nodes+2)], df4, \
                            df3[range(2,num_of_nodes+2)], df5, df6, \
                            df7[range(2,num_of_nodes+2)], df8, df9, df10, \
                            df21[range(2,num_of_nodes+2)], df22, df23, df24, df25 ], ignore_index=True)

        df_z = pd.DataFrame()
        #print(df)
        i=0
        for asa in range(2,num_of_nodes+2):
            node = df_new[asa]
            z_value = stats.zscore(node)
            df_z.insert(i, asa, z_value)
            i+=1
        
        save_name = "./"+network+"/"+replicate+"/"+topo+".xlsx"
        df_z.to_excel(save_name , header = genes, index = False)
        print("done")
print(df_new)
```
